```python
import numpy as np
import matplotlib.pyplot as plt

# components of a picture height and width
height, width = 400, 300
img = np.zeros([height,width,3], dtype=np.uint8)
img[:,:,0] = 255 # change the first elemnet of the first coulmn
img[:1,:,0] = 255 # change the first elemnt only
img

plt.imshow(img)
plt.show()
```

```python
## Lets make it green  ( it will be yellow because i din't make it back as zero)
img[:,:,1] = 255 # change the first elemnet of the first coulmn
plt.imshow(img)
plt.show()
```

```python
img[:,:,2] = 255 # change the first elemnet of the first coulmn
plt.imshow(img)
plt.show()

# rows, coulmns, color img[:,:,2]
```

Here we generate a random image using randint

```python

# size height width depth (rows)
random_image = np.random.randint(0, 256, (height, width, 3), dtype=np.uint8)
plt.imshow(random_image)
plt.show()
```

```python
# generation to a noise (in eache pixel) intense to it 
new_image = random_image + img
plt.imshow(new_image)
plt.show()
```

```python
## we can use  that to encryption or decription back
new_image = random_image - img
plt.imshow(new_image)
plt.show()
```
